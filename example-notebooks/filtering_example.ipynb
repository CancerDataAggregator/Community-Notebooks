# Find data from TCGA-BRCA project, with donors over the age of 50 with imaging data

```python
q1 = Q('File.associated_project = "TCGA-BRCA"')
q2 = Q('days_to_birth < -50*365')
q3 = Q('identifier.system = "IDC"')

q = q3.From(q1.And(q2))
r = q.run(limit = 20)
```

Below I'm using pandas json_normalize function to flatten the data and observe the files. If you look at the output below you will see that there are `GDC` files.

```python
pd.json_normalize(r, record_path =['ResearchSubject','Diagnosis'],meta=['species','sex','race','ethnicity'],meta_prefix= 'subject_', max_level = 2,errors='ignore')
```

An even closer look revels that there is also `PDC` files.

```python
pd.json_normalize(r, record_path =['File','identifier'],meta=['species','sex','race','ethnicity'],meta_prefix= 'subject_', max_level = 1,errors='ignore')['system'].unique()
```

```python
pd.json_normalize(r, record_path =['File'],meta=['species','sex','race','ethnicity'],meta_prefix= 'subject_', max_level = 1,errors='ignore').info()
```
